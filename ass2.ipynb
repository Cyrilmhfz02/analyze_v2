<h3><font color="red">Function</font></h3>

```python

#creating the function that takes IDs and calculate the distance between the two houses
ids=list(df['id'].values)

def distance(id1,id2):
    if(id1 not in ids):
        return "there is np such id, change the first id"
    
    elif(id2 not in ids):
        return "there is no such id, change the second id"
    
    elif((id1 not in ids) and (id2 not in ids)):
       return "change both ids"

    else:
        #selecting the longitude using loc to locate the longitude according to the id
        lon1 = radians(df.loc[df.id==id1]['long'].values[0])
        lon2 = radians(df.loc[df.id==id2]['long'].values[0])
        
        #selecting the latitude using loc to locate the latitude according to the id
        lat1 = radians(df.loc[df.id==id1]['lat'].values[0])
        lat2 = radians(df.loc[df.id==id2]['lat'].values[0])
      
        # Haversine formula
        dlon = lon2 - lon1
        dlat = lat2 - lat1
        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
 
        c = 2 * asin(sqrt(a))
    
        # Radius of earth in kilometers. Use 3956 for miles
        r = 6371
      
        # calculate the result
        return(c * r)

        

id1=int(input("emter the id of the first house: "))
id2=int(input("enter the id of the second house: "))
print(f"the distance between both of these houses is: {distance(7129300520,1954400510)} km")



     
```

```python
#calculating the distance and the price

def distance_price(id1,id2):
    if isinstance(distance(id1,id2),str):
        return distance(id1,id2)

    else:
        price1=df.loc[df.id==id1].price.values[0]
        price2=df.loc[df.id==id2].price.values[0]

        answer= (
            f"Distance of the two houses: {distance(id1,id2)} km\n"
            f"Price of the first house {price1}$\n"
            f"Price of the second house {price2}$"
        )
        return answer

print(distance_price(1954400510,7129300520))
```

<h5>.at()</h5>

```python
#The .at function is used to select a specific value when you have the exact name of the column and the exact name of the index
#the .at() is used when we have labels on the rows and the columns

#example:
col_df={'lat':[1,4,5],
        'long':[1,2,3]}

#here we are creating a df that have as column the values in col_df
#for the name of the rows we are using index=[] to set up the label for the rows
ex_df = pd.DataFrame(col_df, index=['row1', 'row2', 'row3'])

ex_df.at['row2','lat']
```

```python
#The .iat function is used when we dont have label as indexes
#we will be accessing both column and name through the index
df_ex = {'A': [1, 2, 3], 
         'B': [4, 5, 6]}

df_ex=pd.DataFrame(df)
#we are accessing the value in the second row first column
df_ex.iat[1,0]
```
